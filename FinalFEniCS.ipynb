```python
import numpy as np
import fenics as fe
import matplotlib.pyplot as plt
```

```python
from dolfin import *

c=5
nodes = 80
#Primero, creamos la malla, en este caso es una malla cuadra de 4x4
meshW = fe.RectangleMesh(fe.Point(-2,-2), fe.Point(2,2),nodes,nodes)

function_space = fe.FunctionSpace(meshW, "Lagrange", 1)

# Time variables
dt = 0.004; t = 0; T = 4

#condición de frontera
uW_on_boundary = fe.Constant(0.0)
bc = fe.DirichletBC(function_space, uW_on_boundary, "on_boundary")
#condicion inicial
initial_pulse = np.sin(c * 10 * t)

# Previous and current solution
u1= fe.interpolate(uW_on_boundary, function_space)
u0= fe.interpolate(uW_on_boundary, function_space)

# Variational problem at each time
u = fe.TrialFunction(function_space)
v = fe.TestFunction(function_space)

a = u*v*fe.dx + dt*dt*c*c*fe.inner(fe.grad(u), fe.grad(v))*fe.dx
L = 2*u1*v*fe.dx-u0*v*fe.dx


A, b = fe.assemble_system(a, L, bc)

gif_file = fe.File('output2D/sol.pvd')
u= fe.Function(function_space)
while t <= T:
    A, b = fe.assemble_system(a, L, bc)
    delta = fe.PointSource(function_space, fe.Point(0, 0), np.sin(c * 10 * t) )
    delta.apply(b)
    fe.solve(A, u.vector(), b)
    u0.assign(u1)
    u1.assign(u)
    t += dt
    gif_file << (u,t)
    fe.plot(u)

fe.plot(u)
plt.title("Wave equation 2D")
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

```python
elements = 32
mesh = fe.UnitIntervalMesh(elements)

lagrange_polynomial_space_first_order = fe.FunctionSpace(mesh,"Lagrange",1)

#initial boundary conditions
# The value of the solution on the boundary
u_on_boundary = 0.0

# The homogeneous Dirichlet Boundary Condition
boundary_condition = fe.DirichletBC(lagrange_polynomial_space_first_order,u_on_boundary, "on_boundary")

# The initial condition, u(t=0, x) = sin(pi * x)
initial_condition = fe.Expression("sin(3.141 * x[0])",degree=1)

# Discretize the initial condition
u_old = fe.interpolate(initial_condition, lagrange_polynomial_space_first_order)
plt.figure()
fe.plot(u_old, label="t=0.0")

# The time stepping of the implicit Euler discretization (=dt)
time_step_length = 0.1

# Create the Finite Element Problem
u_trial = fe.TrialFunction(lagrange_polynomial_space_first_order)
v_test = fe.TestFunction(lagrange_polynomial_space_first_order)

weak_form_residuum = (u_trial * v_test * fe.dx + time_step_length * fe.dot(fe.grad(u_trial), fe.grad(v_test)) * fe.dx  -
    (u_old * v_test * fe.dx ))

# We have a linear PDE that is separable into a lhs and rhs
weak_form_lhs = fe.lhs(weak_form_residuum)
weak_form_rhs = fe.rhs(weak_form_residuum)

# The function we will be solving for at each point in time
u_solution = fe.Function(lagrange_polynomial_space_first_order)

# time stepping
n_time_steps = 5

time_current = 0.0
for i in range(n_time_steps):
    time_current += time_step_length

    # Finite Element Assembly, BC imprint & solving the linear system
    fe.solve(weak_form_lhs == weak_form_rhs, u_solution,boundary_condition)

    u_old.assign(u_solution)

    fe.plot(u_solution, label=f"t={time_current:1.1f}")

plt.legend()
plt.title("Heat Conduction in a rod with homogeneous Dirichlet BC")
plt.xlabel("x position")
plt.ylabel("Temperature")
plt.grid()
plt.show()
```

```python
t_end = 10.0
dt = 0.1
k = 300
u_in = 20
u_out = -20

xml_file = "helix.xml"
mesh = fe.Mesh(xml_file)
fd = fe.MeshFunction('size_t', mesh, "helix_facet_region.xml");

V = fe.FunctionSpace(mesh, 'P', 1)

bc1 = fe.DirichletBC(V, fe.Constant(u_in), fd, 3)
bc2 = fe.DirichletBC(V, fe.Constant(u_out), fd, 2)
bc = [bc1, bc2]

u = fe.TrialFunction(V)
v = fe.TestFunction(V)
u_n = fe.Function(V)

F = u*v*fe.dx + dt*k*fe.dot(fe.grad(u), fe.grad(v))*fe.dx - u_n*v*fe.dx
a, L = fe.lhs(F), fe.rhs(F)

u = fe.Function(V)
t = 0
vtkfile = fe.File('output/output.pvd')

num_steps = int(t_end/dt)
for n in range(num_steps):
    t += dt
    fe.solve(a == L, u, bc)
    u_n.assign(u)
    vtkfile << (u, t)
```
